```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
pokemon_data = pd.read_csv('Pokemon.csv')
```

```python
pokemon_data.head(10)
```

```python
pokemon_data.columns = pokemon_data.columns.str.upper()
```

```python
pokemon_data['TYPE 2'].fillna(value='None',inplace=True)
```

```python
pokemon_data.drop_duplicates('#',keep='first',inplace = True)
```

```python
pokemon_data.head(10)
```

```python
print('The unique pokemon types are: ',pokemon_data['TYPE 1'].unique())
```

```python
print('MAX HP Pokemon: ',pokemon_data['HP'].idxmax())
```

```python
pokemon_data['#'].count()
```

```python
print(pokemon_data.iloc[0])
```

```python
pokemon_data.sort_values(by='TOTAL',ascending=False)
```

```python
sns.countplot(pokemon_data['GENERATION'])
```

```python
sns.jointplot(x='SP. ATK', y='SP. DEF',data=pokemon_data,kind='hex',height=7)
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(pokemon_data[pokemon_data['TYPE 2']!='None'].groupby(['TYPE 1', 'TYPE 2']).size().unstack(),annot=True,cmap='YlGnBu')
```

```python
plt.figure(figsize=(20,15))
sns.swarmplot(x='TYPE 1',y='TOTAL',data=pokemon_data,hue='LEGENDARY')
plt.axhline(pokemon_data['TOTAL'].mean(),color='red',linestyle='dashed')
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x='GENERATION',y='TOTAL',data=pokemon_data)
```

```python
plt.figure(figsize=(7,7))

legpoke = pokemon_data[pokemon_data['LEGENDARY']==True]
nonlegpoke = pokemon_data[pokemon_data['LEGENDARY']==False]

legcount = [legpoke['#'].count(),nonlegpoke['#'].count()]

plt.pie(legcount,
        labels= ['Legendary', 'Non Legendary'],
        autopct ='%1.1f%%', 
        colors=['aqua','grey'],
        startangle = 0,
        explode=(0.05,0.05))
```
